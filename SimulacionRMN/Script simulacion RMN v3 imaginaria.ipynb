**Creacion del eje:** Cada atomo de H que es puesto en el campo del RMN (wo) decae con una frecuencia especifica (w) que depende de su ambiente quimico. Los espectros reales se obtienen en el dominio del tiempo y se grafica la intensidad de cada nucleo no en termino de w sino en terminos de chemical shift (delta) que es la relacion de w de una w de referencia. Esto logra que independientemente del campo, el desplazamiento quimico de un atomo para una molecula dada es siempre constante.

Para poder graficar los datos de los espectros individuales subidos en formato jdx se debe crear un eje en ppm. Si se utilizasen espectros en FID, o se simulasen esto deberia modificarse acorde al tipo de archivo de espectros individuales utilizados.

```python
# Creacion de un eje en ppm

def jdx_num(d, key, default=None, cast=float):
    v = d.get(key, default)
    if isinstance(v, list):
        v = v[0] if v else default
    try:
        return cast(v)
    except Exception:
        return default

d0 = standardsDictionary[0]

firstx = jdx_num(d0, 'FIRSTX', 10.0, float)
lastx  = jdx_num(d0, 'LASTX', 0.0, float)
npts   = jdx_num(d0, 'NPOINTS', len(standardsSpec[0]), int)

ppm_axis = np.linspace(firstx, lastx, npts)

# región de señal fija
SIG_START, SIG_END = 10000, 56000 #debe chequearse si estos puntos estan ok respecto a lo que se desea

# Chequeo de longitudes
assert all(len(s) == npts for s in standardsSpec), "Hay espectros con distinta longitud."
```

```python
# Descriptores de concentraciones desde el perfil metabolico

#Función para decidir la inclusión de un compuesto en la mezcla según su probabilidad de ausencia
def entra_en_mezcla(p_ausencia):
    return np.random.rand() > float(p_ausencia)

# Función para muestrear la concentración de un compuesto según distribución estadística (Normal, Uniforme o Triangular
def sample_conc(media, desvio, distribucion):
    m, s = float(media), float(desvio)
    dist = str(distribucion).strip().upper()

    if dist.startswith("N"):
        return max(0.0, np.random.normal(m, s))

    if dist.startswith("R"):
        w = s * np.sqrt(12.0)
        a, b = max(0.0, m - w/2), max(m + w/2, m - w/2 + 1e-12)
        return np.random.uniform(a, b)

    if dist.startswith("T"):
        h = s * np.sqrt(6.0)
        a, c = max(0.0, m - h), max(m + h, m - h + 1e-12)
        return np.random.triangular(a, m, c)  # <-- antes decía np.triangular

    return m
```

```python
# Función para aplicar un corrimiento aleatorio en puntos al espectro

max_shift = 30         # corrimiento máximo (en puntos)

def corrimiento_aleatorio(spec, max_shift):
    shift_pts = int(round(np.random.uniform(-1, 1) * max_shift))
    return ng.process.proc_base.cs(spec, shift_pts)
```

```python
# Kernels para deformación de picos en el campo de la frecuencia. Se podria hacer desde el FID si los espectros infividuales estuviesen en FID o se puede volver a la FID desde los espectros.

#es necesario analizar muy bien este item, jugar con todos los parametros para generar deformaciones realistas y buscar otros script para generar estas deformaciones. Las deformaciones de picos no pueden ser resueltas por los sofwares de RMN durantes el procesamiento

def deformar_picos(spec,
                   gauss_sigma_range=(0.0, 2.0),   # ensanchamiento gaussiano en puntos
                   lorentz_gamma_range=(0.0, 2.0), # ensanchamiento lorentziano en puntos
                   asym_prob=0.2,                  # probabilidad de asimetría
                   asym_decay_pts=30,              # decaimiento de la cola (puntos)
                   asym_amp_range=(0.0, 0.10)):    # amplitud relativa de la cola
    y = spec

     #  Gaussiano
    sigma = np.random.uniform(*gauss_sigma_range)
    if sigma > 0:
        L = max(3, int(6*sigma)+1)
        x = np.arange(-(L//2), L//2 + 1)
        g = np.exp(-0.5*(x/sigma)**2)
        g /= g.sum()
        y = np.convolve(y, g, mode='same')

    #  Lorentziano
    gamma = np.random.uniform(*lorentz_gamma_range)
    if gamma > 0:
        L = max(3, int(10*gamma)+1)
        x = np.arange(-(L//2), L//2 + 1)
        l = 1.0 / (1.0 + (x/gamma)**2)
        l /= l.sum()
        y = np.convolve(y, l, mode='same')

    #  Asimetría
    if np.random.rand() < asym_prob and asym_decay_pts > 0:
        amp = np.random.uniform(*asym_amp_range)
        dir_sign = np.random.choice([-1, 1])
        L = max(3, int(5*asym_decay_pts))
        x = np.arange(0, L)
        e = np.exp(-x / float(asym_decay_pts))
        e /= e.sum()

        # núcleo: delta + pequeña cola a un lado
        k = np.zeros(2*L+1)
        k[L] = 1.0 - amp
        if dir_sign > 0:
            k[L+1:L+1+L] = amp * e               # cola a la derecha
        else:
            k[L - L:L] = (amp * e)[::-1]         # cola a la izquierda (equivale a k[:L] = (amp*e)[::-1])

        y = np.convolve(y, k, mode='same')

    return y
```

```python
def aplicar_fase(y_complex, x_ppm, ref_ppm=0.0, prob=1,
                 phi0_range=(-15, 15),     # grados
                 phi1_range=(-3, 3)):    # grados/ppm
    
        # --- sanity checks mínimos ---
    if not (0.0 <= float(prob) <= 1.0):
        raise ValueError("prob debe estar entre 0 y 1.")
    if not np.iscomplexobj(y_complex):
        raise ValueError("y_complex debe ser COMPLEJO (usa hilbert antes).")

    y_complex = np.asarray(y_complex)
    x_ppm     = np.asarray(x_ppm, dtype=float)

    if y_complex.shape != x_ppm.shape:
        raise ValueError("y_complex y x_ppm deben tener la MISMA forma.")

    aplicar = (np.random.rand() < prob)

    # Parámetros de fase
    phi0 = np.random.uniform(*phi0_range) if aplicar else 0.0
    phi1 = np.random.uniform(*phi1_range) if aplicar else 0.0
    phi_rad = np.deg2rad(phi0 + phi1 * (x_ppm - ref_ppm))  # shape (N,)
    y_out = y_complex * np.exp(1j * phi_rad)               # rotación compleja

    return y_out, int(aplicar), float(phi0), float(phi1), float(ref_ppm)
```

```python


# ——— Configuración de porcentajes (suman 1.0) ———
PCT_NO_PHASE        = 0.40  # 30%
PCT_MAX_PEAK        = 0.15  # 20%
PCT_REF_PEAK        = 0.15  # 15%
PCT_PROM_PEAK       = 0.15  # 20% (pico prominente aleatorio)
PCT_CENTER          = 0.15  # 15%

# Si más adelante tenés un pico de referencia conocido (p. ej., TSP=0.0),
# seteá REF_REFERENCE_PPM con ese valor. Por ahora lo dejamos en None.
REF_REFERENCE_PPM = None   # Ejemplo futuro: 0.0  (TSP)  ó alguna otra referencia conocida

def _ppm_center(ppm_axis):
    """Centro del eje (10→0 ⇒ ~5 ppm)."""
    return float((np.max(ppm_axis) + np.min(ppm_axis)) / 2.0)

def _ref_from_max_peak(ppm_axis, y):
    """Pivote = ppm del pico de mayor intensidad absoluta."""
    idx = int(np.argmax(np.abs(y)))
    return float(ppm_axis[idx])

def _ref_from_prominent_random(ppm_axis, y, q=0.90):
    """
    Pivote = un pico 'prominente' elegido al azar entre los puntos
    por encima del cuantil q de |y|. Si no hay suficientes, cae al máximo.
    """
    yabs = np.abs(y)
    thr = np.quantile(yabs, q)
    candidates = np.where(yabs >= thr)[0]
    if candidates.size == 0:
        return _ref_from_max_peak(ppm_axis, y)
    idx = int(np.random.choice(candidates))
    return float(ppm_axis[idx])

def choose_phase_strategy(ppm_axis, y):
    """
    Devuelve: (apply_phase(bool), ref_ppm(float or np.nan), mode_tag(str))
    según las proporciones pedidas.
    """
    u = random.random()
    cut_no   = PCT_NO_PHASE
    cut_max  = cut_no   + PCT_MAX_PEAK
    cut_ref  = cut_max  + PCT_REF_PEAK
    cut_prom = cut_ref  + PCT_PROM_PEAK
    # cut_center = 1.0

    if u < cut_no:
        return False, float('nan'), "no_phase"

    if u < cut_max:
        return True, _ref_from_max_peak(ppm_axis, y), "max_peak"

    if u < cut_ref:
        if REF_REFERENCE_PPM is not None:
            return True, float(REF_REFERENCE_PPM), "ref_peak"
        else:
            # Placeholder activo: si aún no hay referencia, caemos al centro
            return True, _ppm_center(ppm_axis), "ref_peak_fallback_center"

    if u < cut_prom:
        return True, _ref_from_prominent_random(ppm_axis, y, q=0.90), "prominent_peak"

    # Centro del eje
    return True, _ppm_center(ppm_axis), "center"
```

```python
from scipy.signal import hilbert

#  Generacion de espectros por mezcla de compuestos individuales

np.random.seed(2)
random.seed(2)

iterations = 10000      # cuántas mezclas sintéticas generar
standard_scale = 1.0    # factor de intensidad fijo para el estándar

spectra = []
concs = []

REF_PPM = float(np.median(ppm_axis))

phase_applied_list, phi0_list, phi1_list, ref_list = [], [], [], []
phase_mode_list = []

# ——————————————————— Bucle de mezclas ————————————————————
for _ in range(iterations):
    mix_real = np.zeros_like(standardsSpec[0])
    conc_dict = {}

    # sumar compuestos con ausencia, concentración y corrimiento/deformación
    for _, row in master_df.iterrows():
        if not entra_en_mezcla(row["p_ausencia"]):
            continue
        conc = sample_conc(row["media"], row["desvio"], row["distribucion"])
        spec = standardsSpec[int(row["idx"])]

        spec_shifted = corrimiento_aleatorio(spec, max_shift)
        spec_deformed = deformar_picos(spec_shifted)

        mix_real += conc * spec_deformed
        conc_dict[int(row["idx"])] = float(conc)

    # agregar estándar interno
    m = min(len(mix_real), len(v))
    mix_real = mix_real[:m] + standard_scale * v[:m]
    conc_dict["_IS"] = float(standard_scale)

    # ——— COMPLEJIZACIÓN con Hilbert ———
    mix_complex = hilbert(mix_real)  # ✅ COMPLEJO físico

    # Selección de estrategia y pivote
    apply_phase, ref_ppm_sel, mode_tag = choose_phase_strategy(
        ppm_axis[:len(mix_complex)],
        mix_complex
    )

    if not apply_phase:
        applied = False
        phi0_used = 0.0
        phi1_used = 0.0
        ref_used = np.nan
    else:
        mix_complex, applied, phi0_used, phi1_used, ref_used = aplicar_fase(
            mix_complex,
            ppm_axis[:len(mix_complex)],
            ref_ppm=float(ref_ppm_sel),
            prob=1.0
        )

    # Logs
    phase_applied_list.append(int(bool(applied)))
    phi0_list.append(float(phi0_used))
    phi1_list.append(float(phi1_used))
    ref_list.append(float(ref_used) if np.isfinite(ref_used) else np.nan)
    phase_mode_list.append(mode_tag)

    # baseline y ruido en COMPLEJO
    baseline_offset = np.random.uniform(-0.1, 0.1)
    mix_complex = mix_complex + baseline_offset

    noise_sigma = np.random.uniform(0.01, 0.1)
    noise = (np.random.normal(0, noise_sigma, size=mix_complex.shape) +
             1j*np.random.normal(0, noise_sigma, size=mix_complex.shape))
    mix_complex = mix_complex + noise

    # guardar COMPLEJO
    spectra.append(mix_complex)
    concs.append(conc_dict)
```

```python
# --- config de salida ---
out_dir = os.path.join(root_dir, "Graficosv3")
os.makedirs(out_dir, exist_ok=True)

# Qué componente graficar: 'real' | 'imag' | 'abs'
PLOT_MODE = 'real'

def _to_plot(y, mode='real'):
    if mode == 'real':
        return np.real(y)
    elif mode == 'imag':
        return np.imag(y)
    elif mode == 'abs':
        return np.abs(y)
    else:
        raise ValueError("PLOT_MODE debe ser 'real', 'imag' o 'abs'.")

# --- función utilitaria ---
def cut_ppm(x_axis, y, ppm_min, ppm_max, mode='real'):
    """Devuelve x,y recortados en el rango [ppm_min, ppm_max] usando la componente elegida."""
    y = _to_plot(y, mode=mode)
    i0 = int(np.argmin(np.abs(x_axis - ppm_min)))
    i1 = int(np.argmin(np.abs(x_axis - ppm_max)))
    lo, hi = sorted([i0, i1])
    return x_axis[lo:hi], y[lo:hi]

# ventanas de zoom (ppm)
zoom_windows = [(0.0, 2.5), (2.5, 3.5), (3.5, 5.0), (5.0, 7.5), (7.5, 10.0)]

# elegir 4 mezclas aleatorias (sin reemplazo)
mix_idxs = np.random.choice(len(spectra), size=4, replace=False)

# eje para la ventana señal (usa tus constantes si existen, sino todo el eje)
if 'SIG_START' in globals() and 'SIG_END' in globals():
    x_full = ppm_axis[SIG_START:SIG_END]
else:
    x_full = ppm_axis

for k in mix_idxs:
    # carpeta por mezcla
    mix_dir = os.path.join(out_dir, f"mix_{int(k)}")
    os.makedirs(mix_dir, exist_ok=True)

    # ---------- espectro completo ----------
    if 'SIG_START' in globals() and 'SIG_END' in globals():
        y_full = _to_plot(spectra[k][SIG_START:SIG_END], mode=PLOT_MODE)
    else:
        y_full = _to_plot(spectra[k], mode=PLOT_MODE)

    fig, ax = plt.subplots(figsize=(12, 3))
    ax.plot(x_full, y_full)
    ax.invert_xaxis()  # 10 → 0
    ax.set_title(f"Mezcla sintética #{k} (completo) — {PLOT_MODE}")
    ax.set_xlabel("ppm"); ax.set_ylabel("Intensidad")
    fig.tight_layout()
    full_path = os.path.join(mix_dir, f"mix_{int(k)}_completo_{PLOT_MODE}.png")
    fig.savefig(full_path, dpi=300, bbox_inches="tight")
    plt.close(fig)

    # ---------- zooms ----------
    y_all = spectra[k]  # complejo
    for (zmin, zmax) in zoom_windows:
        xz, yz = cut_ppm(ppm_axis, y_all, zmin, zmax, mode=PLOT_MODE)

        fig, ax = plt.subplots(figsize=(12, 3))
        ax.plot(xz, yz)
        ax.invert_xaxis()
        ax.set_title(f"Mezcla #{k} — zoom {zmin}–{zmax} ppm — {PLOT_MODE}")
        ax.set_xlabel("ppm"); ax.set_ylabel("Intensidad")
        fig.tight_layout()

        zname = f"{zmin:.1f}-{zmax:.1f}".replace(".", "p")
        zoom_path = os.path.join(mix_dir, f"mix_{int(k)}_zoom_{zname}_ppm_{PLOT_MODE}.png")
        fig.savefig(zoom_path, dpi=300, bbox_inches="tight")
        plt.close(fig)

    print(f"✅ Guardado: {mix_dir}")

```

```python
# === Exportación: complejo (IA) + solo real (Mnova) ===

# Carpetas de salida
export_dir_complex = os.path.join(root_dir, "Espectros sinteticosv3", "complejo + real")
export_dir_real    = os.path.join(root_dir, "Espectros sinteticosv3", "real")
os.makedirs(export_dir_complex, exist_ok=True)
os.makedirs(export_dir_real, exist_ok=True)

# Detectar cantidad de mezclas (lista vs array)
if isinstance(spectra, (list, tuple)):
    N = len(spectra)
    def get_mix(i): return np.asarray(spectra[i])
else:
    N = spectra.shape[0]
    def get_mix(i): return np.asarray(spectra[i, :])

print("Mezclas a exportar:", N)

def safe_name(s: str) -> str:
    s = str(s).strip()
    return "".join(c if c.isalnum() or c in " -_." else "_" for c in s)

num_saved = 0
for i in range(N):
    y_complex = get_mix(i)                           # <-- espectro COMPLEJO
    x = ppm_axis[:len(y_complex)].astype(float)      # eje ppm

    # nombre base (si tenés mix_names lo usa; si no, Mezcla_###)
    if 'mix_names' in globals() and i < len(mix_names):
        base = safe_name(mix_names[i])
    else:
        base = f"Mezcla_{i:03d}"

    # --- Exportar COMPLEJO para IA ---
    df_c = pd.DataFrame({
        "ppm":  x,
        "real": np.real(y_complex),
        "imag": np.imag(y_complex),
    })
    fname_c = os.path.join(export_dir_complex, f"{base}.csv")
    if os.path.exists(fname_c):
        fname_c = os.path.join(export_dir_complex, f"{base}_{i:03d}.csv")
    df_c.to_csv(fname_c, index=False)

    # --- Exportar SOLO REAL para Mnova ---
    df_r = pd.DataFrame({
        "ppm":      x,
        "intensity": np.real(y_complex)
    })
    fname_r = os.path.join(export_dir_real, f"{base}_real.csv")
    if os.path.exists(fname_r):
        fname_r = os.path.join(export_dir_real, f"{base}_{i:03d}_real.csv")
    df_r.to_csv(fname_r, index=False)

    num_saved += 1

print(f"✅ {num_saved} CSV (complex) guardados en: {export_dir_complex}")
print(f"✅ {num_saved} CSV (real-only) guardados en: {export_dir_real}")

# Listar algunos para verificar
ejemplos_c = sorted([f for f in os.listdir(export_dir_complex) if f.endswith(".csv")])[:5]
ejemplos_r = sorted([f for f in os.listdir(export_dir_real) if f.endswith(".csv")])[:5]
print("Ejemplos complex:", ejemplos_c)
print("Ejemplos real-only:", ejemplos_r)
```
