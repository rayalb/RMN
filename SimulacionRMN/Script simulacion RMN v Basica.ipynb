```python
# Descriptores de concentraciones desde el perfil metabolico

#Funci√≥n para decidir la inclusi√≥n de un compuesto en la mezcla seg√∫n su probabilidad de ausencia
def entra_en_mezcla(p_ausencia):
    return np.random.rand() > float(p_ausencia)

# Funci√≥n para muestrear la concentraci√≥n de un compuesto seg√∫n distribuci√≥n estad√≠stica (Normal, Uniforme o Triangular)
def sample_conc(media, desvio, distribucion):
    m, s = float(media), float(desvio)

    # normaliza el texto a min√∫sculas y quita guiones / espacios
    dist = str(distribucion).strip().lower()
    dist_key = re.sub(r"[\s\-]", "", dist)
   
    if dist.startswith("normal"):
        return max(0.0, np.random.normal(m, s))
    
    if dist_key.startswith("lognormal"):
        if m <= 0 or s <= 0:
            return max(0.0, m)
        var = s**2
        sigma2 = np.log(1.0 + var/(m**2))
        sigma  = np.sqrt(max(1e-12, sigma2))
        mu     = np.log(m) - 0.5*sigma2
        x = np.random.lognormal(mean=mu, sigma=sigma)
        return max(0.0, float(x))

   # --- Gamma ---
    if dist_key.startswith("gamma"):
        if m <= 0 or s <= 0:
            return max(0.0, m)
        k = (m/s)**2          # shape
        theta = (s**2)/m      # scale
        x = np.random.gamma(shape=max(1e-12, k), scale=max(1e-12, theta))
        return max(0.0, float(x))

    if dist.startswith("rectangular"):
        w = s * np.sqrt(12.0)
        a, b = max(0.0, m - w/2), max(m + w/2, m - w/2 + 1e-12)
        return np.random.uniform(a, b)

    if dist.startswith("triangular"):
        h = s * np.sqrt(6.0)
        a, c = max(0.0, m - h), max(m + h, m - h + 1e-12)
        return np.random.triangular(a, m, c)  

    return max(0.0, m)
```

```python
# Funci√≥n para aplicar un corrimiento aleatorio en puntos al espectro

max_shift = 0    # corrimiento m√°ximo (en puntos)

def corrimiento_aleatorio(spec, max_shift):
    shift_pts = int(round(np.random.uniform(-1, 1) * max_shift))
    spec_shifted = ng.process.proc_base.cs(spec, shift_pts)
    return spec_shifted, shift_pts
```

```python
def aplicar_fase(y_complex, x_ppm, ref_ppm=0.0, prob=1,
                 phi0_range=(-15, 15),     # grados
                 phi1_range=(-3, 3)):    # grados/ppm
    
        # --- sanity checks m√≠nimos ---
    if not (0.0 <= float(prob) <= 1.0):
        raise ValueError("prob debe estar entre 0 y 1.")
    if not np.iscomplexobj(y_complex):
        raise ValueError("y_complex debe ser COMPLEJO (usa hilbert antes).")

    y_complex = np.asarray(y_complex)
    x_ppm     = np.asarray(x_ppm, dtype=float)

    if y_complex.shape != x_ppm.shape:
        raise ValueError("y_complex y x_ppm deben tener la MISMA forma.")

    aplicar = (np.random.rand() < prob)

    # Par√°metros de fase
    phi0 = np.random.uniform(*phi0_range) if aplicar else 0.0
    phi1 = np.random.uniform(*phi1_range) if aplicar else 0.0
    phi_rad = np.deg2rad(phi0 + phi1 * (x_ppm - ref_ppm))  # shape (N,)
    y_out = y_complex * np.exp(1j * phi_rad)               # rotaci√≥n compleja

    return y_out, int(aplicar), float(phi0), float(phi1), float(ref_ppm)
```

```python
def deformar_picos(spec,
                   gauss_sigma_range=(0.0, 2.0),   # ensanchamiento gaussiano en puntos
                   lorentz_gamma_range=(0.0, 2.0), # ensanchamiento lorentziano en puntos
                   asym_prob=0.2,                  # probabilidad de asimetr√≠a
                   asym_decay_pts=30,              # decaimiento de la cola (puntos)
                   asym_amp_range=(0.0, 0.10)):    # amplitud relativa de la cola
    y = spec

    #  Gaussiano
    sigma = np.random.uniform(*gauss_sigma_range)
    if sigma > 0:
        L = max(3, int(6 * sigma) + 1)
        x = np.arange(-(L // 2), L // 2 + 1)
        g = np.exp(-0.5 * (x / sigma) ** 2)
        g /= g.sum()
        y = np.convolve(y, g, mode='same')

    #  Lorentziano
    gamma = np.random.uniform(*lorentz_gamma_range)
    if gamma > 0:
        L = max(3, int(10 * gamma) + 1)
        x = np.arange(-(L // 2), L // 2 + 1)
        l = 1.0 / (1.0 + (x / gamma) ** 2)
        l /= l.sum()
        y = np.convolve(y, l, mode='same')

    #  Asimetr√≠a
    asym_applied = 0   # 0 = no aplicada, 1 = aplicada
    amp = 0.0
    dir_sign = 0       # 1 derecha, -1 izquierda, 0 = sin asimetr√≠a

    if np.random.rand() < asym_prob and asym_decay_pts > 0:
        asym_applied = 1
        amp = np.random.uniform(*asym_amp_range)
        dir_sign = np.random.choice([-1, 1])

        L = max(3, int(5 * asym_decay_pts))
        x = np.arange(0, L)
        e = np.exp(-x / float(asym_decay_pts))
        e /= e.sum()

        # n√∫cleo: delta + peque√±a cola a un lado
        k = np.zeros(2 * L + 1)
        k[L] = 1.0 - amp
        if dir_sign > 0:
            k[L+1:L+1+L] = amp * e               # cola a la derecha
        else:
            k[L-L:L] = (amp * e)[::-1]           # cola a la izquierda

        y = np.convolve(y, k, mode='same')

    # üìå par√°metros de deformaci√≥n para loguear en el Excel
    deform_params = {
        "gauss_sigma": float(sigma),
        "lorentz_gamma": float(gamma),
        "asym_applied": int(asym_applied),
        "asym_amp": float(amp),
        "asym_dir": int(dir_sign),
        "asym_decay_pts": int(asym_decay_pts)
    }

    return y, deform_params
```

```python
ppm_axis = standardsSpec[0][:, 0]

# ‚Äî‚Äî‚Äî Configuraci√≥n de porcentajes (suman 1.0) ‚Äî‚Äî‚Äî
PCT_NO_PHASE        = 1  # 30%
PCT_MAX_PEAK        = 0  # 20%
PCT_REF_PEAK        = 0  # 15%
PCT_PROM_PEAK       = 0  # 20% (pico prominente aleatorio)
PCT_CENTER          = 0  # 15%

# Si m√°s adelante ten√©s un pico de referencia conocido (p. ej., TSP=0.0), # sete√° REF_REFERENCE_PPM con ese valor. Por ahora lo dejamos en None.
REF_REFERENCE_PPM = None   # Ejemplo: 0.0  (TSP)  √≥ alguna otra referencia conocida

def _ppm_center(ppm_axis):
    """Centro del eje (10‚Üí0 ‚áí ~5 ppm)."""
    return float((np.max(ppm_axis) + np.min(ppm_axis)) / 2.0)

def _ref_from_max_peak(ppm_axis, y):
    """Pivote = ppm del pico de mayor intensidad absoluta."""
    idx = int(np.argmax(np.abs(y)))
    return float(ppm_axis[idx])

def _ref_from_prominent_random(ppm_axis, y, q=0.90):
    """
    Pivote = un pico 'prominente' elegido al azar entre los puntos
    por encima del cuantil q de |y|. Si no hay suficientes, cae al m√°ximo.
    """
    yabs = np.abs(y)
    thr = np.quantile(yabs, q)
    candidates = np.where(yabs >= thr)[0]
    if candidates.size == 0:
        return _ref_from_max_peak(ppm_axis, y)
    idx = int(np.random.choice(candidates))
    return float(ppm_axis[idx])

def choose_phase_strategy(ppm_axis, y):
    """
    Devuelve: (apply_phase(bool), ref_ppm(float or np.nan), mode_tag(str))
    seg√∫n las proporciones pedidas.
    """
    u = random.random()
    cut_no   = PCT_NO_PHASE
    cut_max  = cut_no   + PCT_MAX_PEAK
    cut_ref  = cut_max  + PCT_REF_PEAK
    cut_prom = cut_ref  + PCT_PROM_PEAK
    # cut_center = 1.0

    if u < cut_no:
        return False, float('nan'), "no_phase"

    if u < cut_max:
        return True, _ref_from_max_peak(ppm_axis, y), "max_peak"

    if u < cut_ref:
        if REF_REFERENCE_PPM is not None:
            return True, float(REF_REFERENCE_PPM), "ref_peak"
        else:
            # Placeholder activo: si a√∫n no hay referencia, caemos al centro
            return True, _ppm_center(ppm_axis), "ref_peak_fallback_center"

    if u < cut_prom:
        return True, _ref_from_prominent_random(ppm_axis, y, q=0.90), "prominent_peak"

    # Centro del eje
    return True, _ppm_center(ppm_axis), "center"
```

```python
from scipy.signal import hilbert

#  Generaci√≥n de espectros por mezcla de compuestos individuales
np.random.seed(2)
random.seed(2)

iterations = 2000
standard_scale = 1.0
start_id   = 1

spectra = []
concs = []

baseline_offset_list = []
noise_sigma_list     = []
per_compound_params = []

# 1) Eje ppm de referencia desde el primer CSV
ppm_axis = standardsSpec[0][:, 0]
REF_PPM  = float(np.median(ppm_axis))

phase_applied_list, phi0_list, phi1_list, ref_list = [], [], [], []
phase_mode_list = []


# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî Bucle de mezclas ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
aplicar_deformacion = False  #ponelo en False para apagar TODA deformaci√≥n

for k in range(iterations):
    mix_id= start_id + k
    mix_real = np.zeros_like(ppm_axis, dtype=float)
    conc_dict = {}

    # sumar compuestos con ausencia, concentraci√≥n y corrimiento/deformaci√≥n
    for _, row in master_df.iterrows():
        if not entra_en_mezcla(row["p_ausencia"]):
            continue
        conc = sample_conc(row["Media"], row["Desv√≠o"], row["Distribuci√≥n recomendada"])
        spec_y = standardsSpec[int(row["idx"])][:, 1]

        spec_shifted, shift_ppm   = corrimiento_aleatorio(spec_y, max_shift)
        # spec_deformed, deform_params = deformar_picos(spec_shifted)
        if aplicar_deformacion:
            spec_deformed, deform_params = deformar_picos(spec_shifted)
        else:
            spec_deformed = spec_shifted.copy()
            deform_params = {}   # o registrar que NO hubo deformaci√≥n

        mix_real += conc * spec_deformed
        conc_dict[int(row["idx"])] = float(conc)

        row_params = {
            "mix_id":    mix_id,
            "idx":       int(row["idx"]),
            "Compuesto": row["Compuesto"],
            "shift_ppm": float(shift_ppm),
        }
        for k, val in deform_params.items():
            row_params[f"deform_{k}"] = float(val)

        per_compound_params.append(row_params)

       
    # agregar est√°ndar interno
    m = min(len(mix_real), len(v))
    mix_real = mix_real[:m] + standard_scale * v[:m]
    conc_dict["_IS"] = float(standard_scale)

    # --- baseline suave ---
    baseline_offset = np.random.uniform(-0.02, 0.02)
    mix_real = mix_real + baseline_offset
    baseline_offset_list.append(float(baseline_offset))

    # --- ruido gaussiano con nivel relativo ---
    rms = np.sqrt(np.mean(mix_real**2)) + 1e-12
    noise_sigma = 0.02 * rms
    mix_real = mix_real + np.random.normal(0, noise_sigma, size=mix_real.shape)
    noise_sigma_list.append(float(noise_sigma))

    # ‚Äî‚Äî‚Äî COMPLEJIZACI√ìN con Hilbert ‚Äî‚Äî‚Äî
    mix_complex = hilbert(mix_real)

    # Selecci√≥n de estrategia y pivote (usa ppm_axis reconstruido)
    apply_phase, ref_ppm_sel, mode_tag = choose_phase_strategy(
        ppm_axis[:len(mix_complex)],
        mix_complex
    )

    if not apply_phase:
        applied = False
        phi0_used = 0.0
        phi1_used = 0.0
        ref_used = np.nan
    else:
        mix_complex, applied, phi0_used, phi1_used, ref_used = aplicar_fase(
            mix_complex,
            ppm_axis[:len(mix_complex)],
            ref_ppm=float(ref_ppm_sel),
            prob=1.0
        )

    # Logs
    phase_applied_list.append(int(bool(applied)))
    phi0_list.append(float(phi0_used))
    phi1_list.append(float(phi1_used))
    ref_list.append(float(ref_used) if np.isfinite(ref_used) else np.nan)
    phase_mode_list.append(mode_tag)

    # guardar COMPLEJO
    spectra.append(mix_complex)
    concs.append(conc_dict)
```

```python
# --- config de salida ---
out_dir = os.path.join(root_dir, "Graficos basicos", "Prueba 16-12-25C")
os.makedirs(out_dir, exist_ok=True)

# Qu√© componente graficar: 'real' | 'imag' | 'abs'
PLOT_MODE = 'real'

def _to_plot(y, mode='real'):
    if mode == 'real':
        return np.real(y)
    elif mode == 'imag':
        return np.imag(y)
    elif mode == 'abs':
        return np.abs(y)
    else:
        raise ValueError("PLOT_MODE debe ser 'real', 'imag' o 'abs'.")

# --- funci√≥n utilitaria ---
def cut_ppm(x_axis, y, ppm_min, ppm_max, mode='real'):
    """Devuelve x,y recortados en el rango [ppm_min, ppm_max] usando la componente elegida."""
    y = _to_plot(y, mode=mode)
    i0 = int(np.argmin(np.abs(x_axis - ppm_min)))
    i1 = int(np.argmin(np.abs(x_axis - ppm_max)))
    lo, hi = sorted([i0, i1])
    return x_axis[lo:hi], y[lo:hi]

# ventanas de zoom (ppm)
zoom_windows = [(0.0, 2.5), (2.5, 3.5), (3.5, 5.0), (5.0, 7.5), (7.5, 10.0)]

# elegir 4 mezclas aleatorias (sin reemplazo)
mix_idxs = np.random.choice(len(spectra), size=4, replace=False)

# eje para la ventana se√±al (usa tus constantes si existen, sino todo el eje)
if 'SIG_START' in globals() and 'SIG_END' in globals():
    x_full = ppm_axis[SIG_START:SIG_END]
else:
    x_full = ppm_axis

for k in mix_idxs:
    # carpeta por mezcla
    mix_dir = os.path.join(out_dir, f"mix_{int(k)}")
    os.makedirs(mix_dir, exist_ok=True)

    # ---------- espectro completo ----------
    if 'SIG_START' in globals() and 'SIG_END' in globals():
        y_full = _to_plot(spectra[k][SIG_START:SIG_END], mode=PLOT_MODE)
    else:
        y_full = _to_plot(spectra[k], mode=PLOT_MODE)

    fig, ax = plt.subplots(figsize=(12, 3))
    ax.plot(x_full, y_full)
    ax.invert_xaxis()  # 10 ‚Üí 0
    ax.set_title(f"Mezcla sint√©tica #{k} (completo) ‚Äî {PLOT_MODE}")
    ax.set_xlabel("ppm"); ax.set_ylabel("Intensidad")
    fig.tight_layout()
    full_path = os.path.join(mix_dir, f"mix_{int(k)}_completo_{PLOT_MODE}.png")
    fig.savefig(full_path, dpi=300, bbox_inches="tight")
    plt.close(fig)

    # ---------- zooms ----------
    y_all = spectra[k]  # complejo
    for (zmin, zmax) in zoom_windows:
        xz, yz = cut_ppm(ppm_axis, y_all, zmin, zmax, mode=PLOT_MODE)

        fig, ax = plt.subplots(figsize=(12, 3))
        ax.plot(xz, yz)
        ax.invert_xaxis()
        ax.set_title(f"Mezcla #{k} ‚Äî zoom {zmin}‚Äì{zmax} ppm ‚Äî {PLOT_MODE}")
        ax.set_xlabel("ppm"); ax.set_ylabel("Intensidad")
        fig.tight_layout()

        zname = f"{zmin:.1f}-{zmax:.1f}".replace(".", "p")
        zoom_path = os.path.join(mix_dir, f"mix_{int(k)}_zoom_{zname}_ppm_{PLOT_MODE}.png")
        fig.savefig(zoom_path, dpi=300, bbox_inches="tight")
        plt.close(fig)

    print(f"‚úÖ Guardado: {mix_dir}")

```
